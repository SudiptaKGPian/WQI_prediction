```python
#Reading the dataset
data=pd.read_excel('GWQ_2010-2018.xlsx')
data.count()
```

```python
#Converting into dataframe
df = pd.DataFrame(data)
df.head(3)
```

```python
#Removing the inconsistent data
df = df.drop(columns=['SITE_ID', 'PROJECT_NAME', 'Arsenic', 
                      'LR. No', 'Turbidity', '%Na', 'RSC', 'SAR', 'SiO2', 'PO4'])
```

```python
#Removing Null values
df1 = df.dropna()
df1.count()
```

```python
df1.head()
```

```python
# Converting below detection limit to zero
df2 = df1.replace(to_replace ="BDL",
                 value = 0)
```

```python
df2.dtypes
```

```python
#data type conversions
df2['TOT_ALKALINITY']=pd.to_numeric(df2['TOT_ALKALINITY'],errors='coerce')
df2['CA']=pd.to_numeric(data['CA'],errors='coerce')
df2['MG']=pd.to_numeric(data['MG'],errors='coerce')
df2['NA']=pd.to_numeric(data['NA'],errors='coerce')
df2['K']=pd.to_numeric(data['K'],errors='coerce')
df2['FE']=pd.to_numeric(data['FE'],errors='coerce')
df2['CARBONATE']=pd.to_numeric(data['CARBONATE'],errors='coerce')
df2['CHLORIDE']=pd.to_numeric(data['CHLORIDE'],errors='coerce')
df2['SULPHATE']=pd.to_numeric(data['SULPHATE'],errors='coerce')
df2['NITRATE']=pd.to_numeric(data['NITRATE'],errors='coerce')
df2['FLUORIDE']=pd.to_numeric(data['FLUORIDE'],errors='coerce')
df2['TDS']=pd.to_numeric(data['TDS'],errors='coerce')

df2.dtypes
```

```python
# Calculating few new useful parameters from the already known features

NA_percent = ((df2.NA + df2.K)/(df2.NA + df2.K + df2.CA + df2.MG))*100
#NA_percent

KR = df2.NA/(df2.CA + df2.MG)
#KR

MAR = df2.MG/(df2.MG + df2.CA)
#MAR

Total_hardness = 2.5*df2.CA + 4.1*df2.MG
#Total_hardness

RSC = ((df2.CARBONATE + df2.BICARBONATE) - (df2.CA + df2.MG))
#RSC

SAR = df2.NA/(np.sqrt((df2.CA + df2.MG)/2))
#SAR

PI = (((df2.NA + df2.K) + np.sqrt(df2.BICARBONATE))*100)/(df2.NA + df2.K + df2.CA + df2.MG)
#PI
```

```python
# Adding new columns with the calculated parameters
df2['%NA'] = NA_percent
df2['KR'] = KR
df2['MAR'] = MAR
df2['Total_hardness'] = Total_hardness
df2['RSC'] = RSC
df2['SAR'] = SAR
df2['PI'] = PI
```
