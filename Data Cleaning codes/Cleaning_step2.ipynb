```python
#Quality index calculations
Qi=[]
Pi = []
for j in range(1,14):
        
    if j == 1:
        ph = 100*((df3.iloc[0][j]-7)/(cal_data.iloc[j-1][1]-7))
        Qi.append(ph)
    else:
        Qi.append(100*((df3.iloc[0][j])/(cal_data.iloc[j-1][1])))

Qi
```

```python
#Weighted Quality

WiQi = []
for k in range(0,len(Qi)):
    WiQi.append(Qi[k]*cal_data.iloc[k][4])

WiQi
```

```python
sumof_WiQi = []

for i in range(0, df3.shape[0]):
    Qi=[]
    WiQi=[]
    for j in range(1,(cal_data.shape[0]+1)):
        
        
        if j == 1:
            Qi.append(100*((df3.iloc[i][j]-7)/(cal_data.iloc[j-1][1]-7)))
        else:
            Qi.append(100*((df3.iloc[i][j])/(cal_data.iloc[j-1][1])))

    for k in range(0,len(Qi)):
        WiQi.append(Qi[k]*cal_data.iloc[k][4])
    
    sumof_WiQi.append(sum(WiQi))


```

```python
sumof_WiQi
```

```python
### Classifying into different grades of water
```

```python
quality = []
grade = []
for i in range(0, len(sumof_WiQi)):
    
    if sumof_WiQi[i]<=25:
        grade.append('A')
        quality.append('Excellent')
        
    if sumof_WiQi[i]>25 and sumof_WiQi[i]<=50:
        grade.append('B')
        quality.append('Good')

    if sumof_WiQi[i]>50 and sumof_WiQi[i]<=75:
        grade.append('C')
        quality.append('Poor')

    if sumof_WiQi[i]>75 and sumof_WiQi[i]<=100:
        grade.append('D')
        quality.append('Very poor')
        
    if sumof_WiQi[i]>100:
        grade.append('E')
        quality.append('Unsuitable')
        
```

```python
### Adding water quality index, grade and quality into the main dataframe
df3['WQI'] = sumof_WiQi
df3['GRADE'] = grade
df3['QUALITY'] = quality
```

```python
fig, ax = plt.subplots(figsize=(15,12))


matrix = np.triu(df3.corr(), 1)

res = sns.heatmap(df3.corr(), annot=True, cmap='Reds', annot_kws={'size': 14}, mask=matrix)

res.set_xticklabels(res.get_xmajorticklabels(), fontsize = 14)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
res.set_yticklabels(res.get_ymajorticklabels(), fontsize = 14)
cbar = res.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=14)
plt.title('Correlation Heatmap', fontsize = 24, fontweight = 'bold' )

#plt.savefig("Heatmap.png", dpi=300)
```
